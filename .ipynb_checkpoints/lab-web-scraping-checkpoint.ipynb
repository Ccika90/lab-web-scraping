```python
# Function to get book details
def get_book_details():
    base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
    books_data = []
    page_number = 1
    
    while True:
        url = base_url.format(page_number)
        response = requests.get(url)
        
        if response.status_code != 200:
            break  # Exit the loop if the page does not exist
    
        soup = BeautifulSoup(response.text, 'html.parser')
        books = soup.find_all('article', class_='product_pod')
    
        for book in books:
            title = book.h3.a['title']  # Book title
            price = book.find('p', class_='price_color').text.strip()  # Book price
            star_rating = book.find('p', class_='star-rating')['class'][1]  # Book rating
            availability = book.find('p', class_='instock availability').text.strip()  # Availability
            
            # Extract category from the book's detail page
            detail_link = book.h3.a['href']
            detail_url = f"https://books.toscrape.com/catalogue/{detail_link}"
            detail_response = requests.get(detail_url)
            detail_soup = BeautifulSoup(detail_response.text, 'html.parser')
            category = detail_soup.find('ul', class_='breadcrumb').find_all('li')[2].text.strip()  # Category
            description = detail_soup.find('meta', attrs={'name': 'description'})['content'].strip()  # Description
            
            # Append the book details to the list
            books_data.append({
                'Title': title,
                'Price': price,
                'Rating': star_rating,
                'Availability': availability,
                'Category': category,
                'Description': description
            })
        
        page_number += 1
        
    return books_data

# Main execution
if __name__ == "__main__":
    books_data = get_book_details()
    
    # Create a DataFrame
    df = pd.DataFrame(books_data)
    
    # Display the DataFrame
    print(df)
```

```python
# Rename the column 'Category' to 'Genre'
df.rename(columns={'Category': 'Genre'}, inplace=True)

# Display the updated DataFrame
print(df)
```

```python
# Replace the incorrect price format with the correct one
df['Price'] = df['Price'].replace('Â£', '£', regex=True)

# Save the updated DataFrame back to an Excel file (optional)
#excel_df.to_excel('updated_books_data.xlsx', index=False)

# Display the updated DataFrame
print(df)
```

```python
# Filter the DataFrame based on the specified price condition
filtered_df = df[df['Price'] <= '£20.00']

# Display the filtered DataFrame
print(filtered_df)
```

```python
# Filter the DataFrame based on the specified price condition
final_df = filtered_df[filtered_df['Rating'] <= 'Four']

# Display the filtered DataFrame
print(final_df)
```

```python
final_df.reset_index(drop=True, inplace=True)
final_df.tail()
```
